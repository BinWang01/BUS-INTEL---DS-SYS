```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/PacktWorkshops/The-pandas-Workshop/master/Chapter08/Data/PUF2020final_v1coll.csv'
df = pd.read_csv(url)
```

```python
df = df[["REGION", "SQFT", "BEDROOMS", "PRICE"]]
```

```python
df.head(10)
```

```python
df.PRICE.plot(kind = 'hist');
```

```python
df.SQFT.plot(kind = 'hist');
```

```python
df.plot(kind='scatter', y = 'PRICE', x = 'SQFT');
```

```python
df.boxplot(by='BEDROOMS', column='PRICE');
```

```python
df.plot(kind='scatter', x = 'BEDROOMS', y = 'PRICE');
```

```python
df.boxplot(by='REGION', column='PRICE');
```

```python
df.plot(kind='scatter', x = 'REGION', y = 'PRICE')
plt.show()
```

```python
pd.pivot_table(df,index=['REGION']).PRICE.plot(kind='barh');
```
